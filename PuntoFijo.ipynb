# Método del punto fijo

**Definición.** Un punto fijo de una función $g$ es un número $p$ tal que $g(p) = p$.

Insertar imagen

**Teorema.** 
* Si $g \in C[a,b]$ y $g(x) \in [a,b]$ para toda $x \in [a,b]$, entonces $g$ tiene un punto fijo en $[a,b]$.
* Si además, $g'(x)$ existe en $(a,b)$ y existe una constante positiva $k < 1$ con $$|g'(x)| \leq k, \qquad \text{para toda } \qquad x \in (a,b),$$ entoces el punto fijo en $[a,b]$ es único. 

## Método:

Considere una función $f$ definida en el intevalo $[a,b]$ que satisface las condiciones del teorema, entonces para  aproximar el punto fijo de la función, procedemos como sigue:
1. Escogemos una aproximación inicial $p_0 \in [a,b]$ 
2. Generamos la sucesión $p_n = g(p_{n-1})$ para cada $n \geq 1$.

Entonce la sucesión $\{p_n\}_{n=0}^\infty$ converge a un número $p \in [a,b]$ y, además, este valor $p$ es un punto fijo de $f$.

```python
import matplotlib.pyplot as plt
from numpy import sin, arange

def f(x):
    return x + sin(2*x)

t = arange(1,5,0.01)
plt.plot(t,f(t),t,t)


x=5
for i in range(5):
    print("term = %d \t\t %f" %(i,f(x)))
    x=f(x)

```
