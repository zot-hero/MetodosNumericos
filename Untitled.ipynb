```python
#Programa que calcula la raíz de una función por el método de bisección
import numpy as np

def f(x):
	return x**3+4*x**2-10

a=1
b=2
Tol=1e-10
N_0=1000

F_a=f(a)
F_b=f(b)
contador=0

while contador<N_0:
	r=a+(b-a)/2
	F_r=f(r)
	contador = contador+1
	print("La raíz es: ",r)
	if F_r==0  or b-a<Tol:
		break
	else:
		if F_a*F_r>0:
			a=r
		else:
			b=r



print("La aproximación a la raíz es: ",r)
print("Contador : ",contador)
	

import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optimize

print(optimize.bisect(f,1,2, xtol = 1e-10,full_output = True))
```
